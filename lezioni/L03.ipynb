# La matrice delle tabelline

Costruiamo una matrice (lista di liste) di 10 righe e 10 colonne in cui l'$n$-esima riga (lista) contenga la tabellina del numero $n+1$ (ossia una lista contenente i primi 10 multipli di $n+1$).

Nella versione più semplice che realizzeremo, usiamo una funzione `tabellina_del(n)` che 
restituisce la tabellina del numero `n` (ossia una lista contenente i primi 10 multipli di `n`), e poi usiamo questa funzione per costruire la matrice delle tabelline.

```python
def tabellina_del(n):
  tabellina = []
  for i in range(1, 11):
    tabellina.append(n * i)
  return tabellina

tabellina_del(3)
```

```python
tabelline = []
for n in range(1, 11):
  tabelline.append(tabellina_del(n))

tabelline
```

Una versione un po' più compatta prevederebbe l'uso di due `for` annidati

```python
tabelline = []
for n in range(1, 11):
  tabellina = []
  for i in range(1, 11):
    tabellina.append(n * i)
  tabelline.append(tabellina) 

tabelline
```

La costruzione di una lista a partire da un ciclo `for` è così comune che Python fornisce una sintassi speciale per farlo in modo più compatto, chiamata *list comprehension* (comprensione di liste), la funzione `tabellina_del(n)` potrebbe essere riscritta come

```python
n = 3 

[i * n for i in range(1, 11)]
```

Usando di nuovo una list comprehension, potremmo anche costruire la matrice delle tabelline in un'unica riga di codice

```python
[[i * n for i in range(1, 11)] for n in range(1, 11)]
```

# Anagrammi

Un modo semplice per ottenere gli anagrammi conteunti in un elenco di parole è basato sull'idea di *firma* di una parola, intesa come la parola ottenuta ordinando le lettere in ordine alfabetico

```python
def firma(parola):
  lettere = list(parola)
  lettere.sort()
  return ''.join(lettere)

firma('cena') == firma('acne')
```

Appare evidente che due parole sono anagrammi se e solo se hanno la stessa firma.

Ora raccogliamo un *insieme* di parole dal repository [napolux/paroleitaliane](https://github.com/napolux/paroleitaliane) facendo uso della funzione `urlopen` del modulo `urllib.request` 

```python
from urllib.request import urlopen

url = 'https://raw.githubusercontent.com/napolux/paroleitaliane/refs/heads/main/paroleitaliane/60000_parole_italiane.txt'

parole = set([parola.decode('utf-8').strip() for parola in urlopen(url)])

len(parole)
```

Ora costruiamo un *dizionario* che abbia per chiavi le firme delle parole e per valori l'*insieme* di parole che hanno quella firma (ossia l'insieme di anagrammi)

```python
firma_a_anagrammi = dict()

for parola in parole:
  f = firma(parola)
  if f not in firma_a_anagrammi:
    firma_a_anagrammi[f] = set()
  firma_a_anagrammi[f].add(parola)

len(firma_a_anagrammi)
```

Ora ricaviamo da questo dizionario una *lista* dei suoi valori che contengono più di una parola, ossia la lista degli anagrammi veri e propri

```python
lista_anagrammi = [anagrammi for anagrammi in firma_a_anagrammi.values() if len(anagrammi) > 1]

len(lista_anagrammi)
```

Ora ordiniamo questa lista di insiemi in ordine decrescente rispetto al numero di elementi di ciascun insieme. 

Una prima strategia per ottenere questo ordinamento è produrre una lista di coppie in cui il primo elemento sia il numero di elementi e il secondo l'insieme; ordinando questa nuova lista (lessicograficamente) ed eliminando poi il primo elemento otterremo il risultato cercato

```python
lista_anagrammi_con_lunghezza = [(len(anagrammi), anagrammi) for anagrammi in lista_anagrammi]
lista_anagrammi_con_lunghezza.sort(reverse=True)
lista_anagrammi_ordinati = [anagrammi for lunghezza, anagrammi in lista_anagrammi_con_lunghezza]

for anagrammi in lista_anagrammi_ordinati[:10]:
  print(', '.join(anagrammi))
```

Un approccio alternativo consiste nell'usare il parametro `key` della funzione `sort` per specificare che l'ordinamento deve essere fatto in base alla lunghezza degli insiemi di anagrammi

```python
lista_anagrammi_ordinati = lista_anagrammi.copy()
lista_anagrammi_ordinati.sort(key=len, reverse=True)

for anagrammi in lista_anagrammi_ordinati[:10]:
  print(', '.join(anagrammi))
```
